```python
def get_tweetsdf(topic):
  tweetList = []
  userList = []
  likesList = []
  datetimeList = []
  locationList = []
  hashtags = []

  query  = "modi"
  #query which is running = "(place:7929cea6bd5b32bd AND -is:retweet -has:hashtags virat)"
  # query which is running = "(virat -filter:retweets)"
  cursor = tweepy.Cursor(api.search_tweets,q="(virat) -has:hashtags" ,lang ="en", result_type='recent').items(50) #exculde = retweets
  for t in cursor:
    tweetList.append(t.text)
    userList.append(t.user.name)
    likesList.append(t.favorite_count)
    locationList.append(t.user.location)
    datetimeList.append(t.created_at)
    hashtags.append(t.entities["hashtags"])
  
  df =  pd.DataFrame({"User Name":userList,"Tweets":tweetList,"Likes":likesList,
                      "Date Time":datetimeList,"Location":locationList , "hashtags": hashtags})
  return df
```
